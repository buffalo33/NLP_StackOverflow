```python
# print(df)

# Tag remover goes first, spacy goes second
# any non-word character

def remove_tags(string):
    # result = re.sub('<.*?>','',string)
    try:
        result = re.sub(r"(, '[\W\.]')",r"",string)
        return result
    except:
#         print(string)
        return "invalid sentence"

text_df = pd.concat( [df_hf['Title'].apply(lambda x: remove_tags(x) ), df_hf['BodyMarkdown'].apply(lambda x: remove_tags(x) )] , axis=1)

#text_df

print(text_df)
```

## Encode global non sequential / POS features

```python
#tokenise
df_hf['tokenized_title'] = df_hf['Title'].str.lower().apply(nltk.word_tokenize)
df_hf['tokenized_body'] = df_hf['BodyMarkdown'].str.lower().apply(nltk.word_tokenize)
```

Pos tagging tokens

```python
df_hf['tagged_title'] = df_hf['tokenized_title'].apply(nltk.pos_tag)
df_hf['tagged_body'] = df_hf['tokenized_body'].apply(nltk.pos_tag)
```

```python
# # 0 if open, 1 if closed
df_hf['y'] = 1-(df_hf.OpenStatus == 'open').astype(int)

# # Encode 'post creation date' as its age in days
df_hf.PostCreationDate = pd.to_datetime(df_hf.PostCreationDate)
origion_date = max(df_hf.PostCreationDate)
df_hf['PostAge'] = (origion_date-df_hf.PostCreationDate).dt.days

# # Encode OwnerCreationDate as account age in days
df_hf.OwnerCreationDate = pd.to_datetime(df_hf.OwnerCreationDate)
origion_date = max(df_hf.OwnerCreationDate)
df_hf['AccountAge'] = (origion_date-df_hf.OwnerCreationDate).dt.days

# # Encode title and body text as their length
df_hf['TitleLength'] = df_hf.apply(lambda x: len(x['Title'].split()), 1)
df_hf['BodyLength'] = df_hf.apply(lambda x: len(x['BodyMarkdown'].split()), 1)

# # Number of tags a post has (between 0-5)
df_hf['TotalTags'] = df_hf.apply(lambda x: sum([not pd.isnull(x[f'Tag{a}']) for a in range(1, 6)]), 1)
```

Extract the nouns and the counts from the text

```python
df_hf['nouns_title'] = df_hf['tagged_title'].apply(lambda x: [word for word, tag in x if tag in ['NN', 'NNS', 'NNP', 'NNPS']])
```

```python
df_hf['nouns_body'] = df_hf['tagged_body'].apply(lambda x: [word for word, tag in x if tag in ['NN', 'NNS', 'NNP', 'NNPS']])
```

```python
df_hf['count_nouns_body'] = df_hf['nouns_body'].apply(len)
df_hf['count_nouns_title'] = df_hf['nouns_title'].apply(len)
```

Extract the verbs and the counts from the text

```python
df_hf['verbs_title'] = df_hf['tagged_title'].apply(lambda x: [word for word, tag in x if tag in ['VB', 'VBD', 'VBG', 'VBN', 'VBP','VBZ']])
df_hf['verbs_body'] = df_hf['tagged_body'].apply(lambda x: [word for word, tag in x if tag in ['VB', 'VBD', 'VBG', 'VBN', 'VBP','VBZ']])
df_hf['count_verbs_body'] = df_hf['verbs_body'].apply(len)
df_hf['count_verbs_title'] = df_hf['verbs_title'].apply(len)
```

Extract the adjectifs and the counts from the text

```python
df_hf['adj_title'] = df_hf['tagged_title'].apply(lambda x: [word for word, tag in x if tag in ['JJR', 'JJS', 'JJ']])
df_hf['adj_body'] = df_hf['tagged_body'].apply(lambda x: [word for word, tag in x if tag in ['JJR', 'JJS', 'JJ']])
df_hf['count_adj_body'] = df_hf['adj_body'].apply(len)
df_hf['count_adj_title'] = df_hf['adj_title'].apply(len)
```

Extract the exclamation/interrogation and the counts from the text

```python
df_hf['nb_exclamations_body'] = df_hf['BodyMarkdown'].str.count('!')
```
